```python
def get_team_agianst_high(name):
    viratmatches=deliveries[deliveries['batsman']==name]
    return viratmatches.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
get_team_agianst_high('SK Raina')
```
